### Q1: Which team scored the most points when playing at home? 

```python
# Attempt Question 1 here
#Which team scored the most points when playing at home?

sql_text_query = '''
SELECT Team.team_long_name, SUM(Match.home_team_goal)
FROM [Team] 
JOIN [Match] ON Team.team_api_id=Match.home_team_api_id
GROUP BY Team.team_long_name
ORDER BY SUM(Match.home_team_goal) DESC LIMIT 5;
'''

pd.read_sql_query(sql_text_query,conn)

#'SELECT Team.team_long_name, \ 
#Match.home_team_goal FROM Team JOIN Match ON Team.id=Match.id;'
```

```python
# From Stackoverflow --- https://stackoverflow.com/questions/15763965/how-can-i-calculate-the-median-of-values-in-sqlite

#SELECT AVG(x)
#FROM (SELECT x
#      FROM MyTable
#      ORDER BY x
#      LIMIT 2
#      OFFSET (SELECT (COUNT(*) - 1) / 2
#              FROM MyTable))


sql_text_query = '''

SELECT ROUND(AVG(home_team_goal), 2) AS median_item_count  
FROM (  SELECT home_team_goal
        FROM ( SELECT home_team_goal,
           COUNT(*) OVER() AS row_count,
           ROW_NUMBER() OVER(ORDER BY home_team_goal=away_team_goal) AS row_number
    FROM [Match]
    WHERE home_team_goal=away_team_goal <> 0
  ) x
WHERE row_number IN ((row_count + 1)/2, (row_count + 2)/2)  
) y
;
'''
pd.read_sql_query(sql_text_query,conn)

#OFFSET (SELECT (COUNT(*) - 1) / 2 FROM [Match])
```
